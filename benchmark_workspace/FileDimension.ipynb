# Analysis of perfomance wrt File Size

```python
from b_tools.preprocess import get_standardized_measures
from b_tools.analysis import compute_performances
```

```python
# FILES
FILE_SIZES = ['0', '20', '40', '60', '80', '100']

DATA = dict()
for F_SIZE in FILE_SIZES:
    DATA[F_SIZE] = {}
    DATA[F_SIZE]['SEQ'], DATA[F_SIZE]['THR'], DATA[F_SIZE]['FF'] = compute_performances(F_SIZE)    
```

```python
import matplotlib.pyplot as plt
```

## Execution time

```python
plt.figure(figsize=(15,6))
plt.title("Execution time")
# plt.yscale('log')

plt.xlabel("File size")
plt.ylabel("Time [ns]")

for D in DATA.values():
    plt.plot([i for i in range(0, len(D['SEQ']))], [D['SEQ'].median() for _ in range(0, len(D['SEQ']))], '-xb', label="Sequential")
    #plt.plot(FILE_SIZES, [a['TOTAL'].median() for a in D.values()], '-xb', label="Sequential")
    # Plot the median time for Sequential version
    
    #plt.plot(FILE_SIZES, [a['THR']['TOTAL'].median() for a in D.values()], '-xr', label="THR")
    #plt.plot(FILE_SIZES, [a['FF']['TOTAL'].median() for a in D.values()], '-xg', label="FF")

plt.legend()
```
