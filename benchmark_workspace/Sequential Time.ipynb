```python
from b_tools.preprocess import get_standardized_measures
import matplotlib.pyplot as plt
```

```python
data_50 = get_standardized_measures("results/sequential_50.csv")
data_100 = get_standardized_measures("results/sequential_100.csv")
```

```python
x = range(0, len(data_50['TOTAL']))
y50 = data_50['TOTAL']
y100 = data_100['TOTAL']
```

```python
plt.title("Execution time")
# plt.yscale('log')
plt.plot(x, y50, '-xb')
plt.plot(x, y100, '-xr')

plt.xticks(x)
plt.xlabel("Iteration index")
plt.ylabel("Time [ns]")
plt.legend(['50M', '100M'])

plt.show()
```
