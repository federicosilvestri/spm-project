# Analysis of perfomance wrt File Size

```python
from b_tools.preprocess import get_standardized_measures
from b_tools.analysis import compute_performances
```

```python
# FILES
FILE_SIZES = ['1', '2', '3', '5', '50']

DATA = dict()
for F_SIZE in FILE_SIZES:
    DATA[F_SIZE]['SEQ'], DATA[F_SIZE]['THR'], DATA[F_SIZE]['FF'] = compute_performances(F_SIZE)    
```

```python
DATA
```

```python
import matplotlib.pyplot as plt
```

## Execution time

```python
plt.figure(figsize=(15,6))
plt.title("Execution time")
# plt.yscale('log')

plt.xlabel("File size")
plt.ylabel("Time [ns]")

# Plot the median time for Sequential version
plt.plot(FILE_SIZES, [a['SEQ']['TOTAL'].median() for a in DATA.values()], '-xb', label="Sequential")
plt.plot(FILE_SIZES, [a['THR']['TOTAL'].median() for a in DATA.values()], '-xr', label="THR")
plt.plot(FILE_SIZES, [a['FF']['TOTAL'].median() for a in DATA.values()], '-xg', label="FF")

plt.legend()
```
