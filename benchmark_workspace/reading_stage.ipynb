```python
import pandas as pd


df = pd.read_csv("result.csv", delimiter=",", dtype=int)
```

```python
import matplotlib.pyplot as plt

plt.title("Execution time")
#plt.yscale('log')
plt.plot(x, y, '-bo')

plt.xticks(range(0, 21))
plt.xlabel("P")
plt.ylabel("Time [us]")

plt.show()
```
